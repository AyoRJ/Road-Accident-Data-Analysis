```python
import sqlite3

# Connects to the database
conn = sqlite3.connect("single_vehicle_accidents.db")
cursor = conn.cursor()

# Creates the ccidents table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Accidents (
    Accident_Index TEXT PRIMARY KEY, -- Unique identifier for accidents and it is also unique
    Day_of_Week TEXT,
    Date DATE, 
    Time TEXT,
    Year INTEGER CHECK (Year IN (2015, 2016)), -- Only the years 2015 or 2016 can be
    Local_Authority_District TEXT,
    Longitude REAL,
    Latitude REAL,
    Light_Conditions TEXT,
    Urban_or_Rural_Area TEXT,
    Weather_Conditions TEXT,
    Road_Surface_Conditions TEXT,
    Road_Type TEXT,
    Speed_limit INTEGER CHECK (Speed_limit > 0), -- Speed limit must be above 0
    Accident_Severity TEXT,
    Number_of_Casualties INTEGER,
    Junction_Detail TEXT
);
""")

# Creates the Vehicles table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Vehicles (
    Vehicle_ID INTEGER PRIMARY KEY AUTOINCREMENT, -- Auto-incremented primary key
    Accident_Index TEXT NOT NULL, -- Foreign key to Accidents and there must bee an accident_index
    Make TEXT,
    Age_of_Vehicle REAL, 
    Engine_Capacity_CC REAL, -- Changed from exact name Engine_Capacity_.CC. to avoid issues with syntx
    Vehicle_Manoeuvre TEXT,
    UNIQUE (Accident_Index), -- Makes sure one vehicle per accident
    FOREIGN KEY (Accident_Index) REFERENCES Accidents (Accident_Index) ON DELETE CASCADE
);
""")
# Create the drivers table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Drivers (
    Driver_ID INTEGER PRIMARY KEY AUTOINCREMENT, -- Auto-incremented primary key as it is unique
    Vehicle_ID INTEGER NOT NULL, -- Foreign key to Vehicles
    Age_Band_of_Driver TEXT,
    Sex_of_Driver TEXT,
    UNIQUE (Vehicle_ID), -- Ensures one driver to one vehicle
    FOREIGN KEY (Vehicle_ID) REFERENCES Vehicles (Vehicle_ID) ON DELETE CASCADE
);
""")

# Commit changes
conn.commit()
print("Database has been created.")
```

```python
import sqlite3
import pandas as pd

# Loads the dataset
dataset = pd.read_csv("single_vehicle_accidents.csv")
# Defines the accidents table columns
accidents_columns = [
    "Accident_Index", "Day_of_Week", "Date", "Time", "Year", "Local_Authority_(District)",
    "Longitude", "Latitude", "Light_Conditions", "Urban_or_Rural_Area", "Weather_Conditions",
    "Road_Surface_Conditions", "Road_Type", "Speed_limit", "Accident_Severity",
    "Number_of_Casualties", "Junction_Detail"
]
# Rename columns to match database schema
dataset.rename(columns={
    "Engine_Capacity_.CC.": "Engine_Capacity_CC",  # These two changes will be relevant for the vehicles table
    "make": "Make"                                
}, inplace=True)
# Creates the dataframe for accidents
accidents_df = dataset[accidents_columns].drop_duplicates(subset=["Accident_Index"])

# This makes it Connect to the database
conn = sqlite3.connect("single_vehicle_accidents.db")
cursor = conn.cursor()

# populates the data into the Accidents table
accidents_data = accidents_df.to_dict(orient="records")
cursor.executemany("""
INSERT INTO Accidents (
    Accident_Index, Day_of_Week, Date, Time, Year, Local_Authority_District,
    Longitude, Latitude, Light_Conditions, Urban_or_Rural_Area, Weather_Conditions,
    Road_Surface_Conditions, Road_Type, Speed_limit, Accident_Severity,
    Number_of_Casualties, Junction_Detail
) VALUES (
    :Accident_Index, :Day_of_Week, :Date, :Time, :Year, :Local_Authority_(District),
    :Longitude, :Latitude, :Light_Conditions, :Urban_or_Rural_Area, :Weather_Conditions,
    :Road_Surface_Conditions, :Road_Type, :Speed_limit, :Accident_Severity,
    :Number_of_Casualties, :Junction_Detail
);
""", accidents_data)

# Commits and closes connection to db
conn.commit()
print("Accidents data has been populated")
```

```python
# Gets the Vehicle_ID and links it with Drivers to establish the one-to-one relationship
#it has as Vehicle_ID was autoincremented so this needs to be established with the drivers table
#as it is a foreign key
vehicles_df_with_ids = pd.read_sql_query("SELECT Vehicle_ID, Accident_Index FROM Vehicles;", conn)
# Define Drivers table columns
drivers_columns = [
    "Accident_Index", "Age_Band_of_Driver", "Sex_of_Driver"
]
# Create drivers dataframe and merges it with the Vehicle_IDs
drivers_df = dataset[drivers_columns]
drivers_df = pd.merge(drivers_df, vehicles_df_with_ids, on="Accident_Index")

# populates the data into the drivers table
drivers_data = drivers_df.to_dict(orient="records")
cursor.executemany("""
INSERT INTO Drivers (
    Vehicle_ID, Age_Band_of_Driver, Sex_of_Driver
) VALUES (
    :Vehicle_ID, :Age_Band_of_Driver, :Sex_of_Driver
);
""", drivers_data)

conn.commit()
print("Drivers data has been populated.")
```

```python
# Define Vehicles table columns names
vehicles_columns = [
    "Accident_Index", "Make", "Age_of_Vehicle", "Engine_Capacity_CC", "Vehicle_Manoeuvre"
]

# Create dataframe for vehicles
vehicles_df = dataset[vehicles_columns].drop_duplicates(subset=["Accident_Index"])
# populates the data into the vehicles table
vehicles_data = vehicles_df.to_dict(orient="records")
cursor.executemany("""
INSERT INTO Vehicles (
    Accident_Index, Make, Age_of_Vehicle, Engine_Capacity_CC, Vehicle_Manoeuvre
) VALUES (
    :Accident_Index, :Make, :Age_of_Vehicle, :Engine_Capacity_CC, :Vehicle_Manoeuvre
);
""", vehicles_data)

conn.commit()
print("Vehicles data has been populated.")
```

```python
#This code allows me to see which columns have missing data as I identified what terms are used to fill where
#missing data is. I also know the amount and the percentage so I will use DATA VISUALISATION to display how much
#missing data will be filtered from my database when querying. This allows me to derive more accurate insights
#for data that I am aware of, so missing values do not skew the data. And I can highlight the effect of
#present data.
```

```python
# this lists all columns to check and the tables that they are in
columns_to_check = [
    ("Accidents", "Accident_Index"),
    ("Accidents", "Accident_Severity"),
    ("Accidents", "Date"),
    ("Accidents", "Day_of_Week"),
    ("Accidents", "Junction_Detail"),
    ("Accidents", "Latitude"),
    ("Accidents", "Light_Conditions"),
    ("Accidents", "Local_Authority_District"),
   
    ("Accidents", "Longitude"),
    ("Accidents", "Number_of_Casualties"),
    ("Accidents", "Road_Surface_Conditions"),
    ("Accidents", "Road_Type"),
    ("Accidents", "Speed_limit"),
    ("Accidents", "Time"),
    
    ("Accidents", "Urban_or_Rural_Area"),
    ("Accidents", "Weather_Conditions"),
    ("Accidents", "Year"),
    ("Drivers", "Age_Band_of_Driver"),
    ("Vehicles", "Age_of_Vehicle"),
    ("Vehicles", "Engine_Capacity_CC"),
    
    ("Vehicles", "Make"),
    ("Drivers", "Sex_of_Driver"),
    ("Vehicles", "Vehicle_Manoeuvre")
]

# creates a dictionary to store the results
results = {}

# allows me to get count all the rows that have the value there
total_rows = {
    "Accidents": pd.read_sql_query("SELECT COUNT(*) as Total FROM Accidents;", conn).iloc[0]["Total"],
    "Vehicles": pd.read_sql_query("SELECT COUNT(*) as Total FROM Vehicles;", conn).iloc[0]["Total"],
    "Drivers": pd.read_sql_query("SELECT COUNT(*) as Total FROM Drivers;", conn).iloc[0]["Total"],
}

# Loops through each column and calculate the amount and the percentage
for table, column in columns_to_check:
    query = f"""
    SELECT COUNT(*) as Count
    FROM {table}
    WHERE {column} IN ('Data missing or out of range', 'Unknown', 'Unallocated');
    """
    count = pd.read_sql_query(query, conn).iloc[0]["Count"]
    percentage = (count / total_rows[table]) * 100
    results[f"{table}.{column}"] = {"Count": count, "Percentage": percentage}

# shows me the results
for column, stats in results.items():
    table_name, column_name = column.split('.')
    print(f"\nTable: {table_name}, Column: {column_name}")
    print(f"Count of rows with specified values: {stats['Count']}")
    print(f"Percentage: {stats['Percentage']:.2f}%")
```

```python
import matplotlib.pyplot as plt

# List all columns along with its associated table from the db so it can check each column/attribute 
columns_to_check = [
    ("Accidents", "Accident_Index"),
    ("Accidents", "Accident_Severity"),
    ("Accidents", "Date"),
    ("Accidents", "Day_of_Week"),
    ("Accidents", "Junction_Detail"),
    ("Accidents", "Latitude"),
    
    ("Accidents", "Local_Authority_District"),
    ("Accidents", "Longitude"),
    ("Accidents", "Number_of_Casualties"),
    ("Accidents", "Road_Surface_Conditions"),
    ("Accidents", "Road_Type"),
    ("Accidents", "Speed_limit"),
    ("Accidents", "Time"),
    
    ("Accidents", "Weather_Conditions"),
    ("Accidents", "Year"),
    ("Drivers", "Age_Band_of_Driver"),
    ("Vehicles", "Age_of_Vehicle"),
    ("Vehicles", "Engine_Capacity_CC"),
    ("Vehicles", "Make"),
    
    ("Vehicles", "Vehicle_Manoeuvre"),
]

# Implements a dictionary to store the percentage of missing values for each column
missing_values_data = {}

# variable to store total rows and counts the amount
total_rows = {
    "Accidents": pd.read_sql_query("SELECT COUNT(*) as Total FROM Accidents;", conn).iloc[0]["Total"],
    "Vehicles": pd.read_sql_query("SELECT COUNT(*) as Total FROM Vehicles;", conn).iloc[0]["Total"],
    "Drivers": pd.read_sql_query("SELECT COUNT(*) as Total FROM Drivers;", conn).iloc[0]["Total"],
}

# Calculates the missing values percentage for each column
for table, column in columns_to_check:
    query = f"""
    SELECT COUNT(*) as Count
    FROM {table}
    WHERE {column} IN ('Data missing or out of range', 'Unknown', 'Unallocated');
    """
    count = pd.read_sql_query(query, conn).iloc[0]["Count"]
    percentage = (count / total_rows[table]) * 100 if total_rows[table] > 0 else 0
    if percentage > 0:
        missing_values_data[f"{table}.{column}"] = percentage

# Visualises missing values as a pie charts as this can show clearly the 
for column, percentage in missing_values_data.items():
    table_name, column_name = column.split('.')
    # Data for pie chart
    labels = ["Missing", "Not Missing"]
    sizes = [percentage, 100 - percentage]
    colors = ["red", "green"]

    # Plots the pie chart and makes sure it is circular
    plt.figure(figsize=(2, 2))
    plt.pie(sizes, labels=labels, colors=colors, autopct="%.2f%%", startangle=90)
    plt.title(f"Percentage of Missing Values in {table_name}.{column_name}")
    plt.axis("equal")  
    plt.show()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connects to the database
conn = sqlite3.connect("single_vehicle_accidents.db")

# Queries to get the Road types with the most accidents and filters out the missing value that is noted as a string
#"unknown"
query_road_accidents = """
SELECT Road_Type, COUNT(*) as Accident_Count
FROM Accidents
WHERE Road_Type != 'Unknown'
GROUP BY Road_Type
ORDER BY Accident_Count DESC;
"""
road_accidents = pd.read_sql_query(query_road_accidents, conn)

# Queries the Road types with the highest accident severity, filtering out 'unknown' as this is how the
#missing value is represented
#it sums the the count of all the different values of accident_severity
query_road_severity = """
SELECT Road_Type, 
       SUM(CASE WHEN Accident_Severity = 'Fatal' THEN 1 ELSE 0 END) as Fatal_Count,
       SUM(CASE WHEN Accident_Severity = 'Serious' THEN 1 ELSE 0 END) as Serious_Count,
       SUM(CASE WHEN Accident_Severity = 'Slight' THEN 1 ELSE 0 END) as Slight_Count
FROM Accidents
WHERE Road_Type != 'Unknown'
GROUP BY Road_Type
ORDER BY Fatal_Count DESC, Serious_Count DESC;
"""
road_severity = pd.read_sql_query(query_road_severity, conn)


# Prints th results
print("Road Types with the Most Accidents:")
print(road_accidents)
print("\nRoad Types with the Highest Accident Severity:")
print(road_severity)

# Visualises the Road types with the most accidents
plt.figure(figsize=(10, 6))
plt.bar(road_accidents["Road_Type"], road_accidents["Accident_Count"], color="skyblue")
plt.title("Road Types with the Most Accidents", fontsize=14)
plt.xlabel("Road Type", fontsize=12)
plt.ylabel("Number of Accidents", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.tight_layout()
plt.savefig("figure111.png", dpi=300, bbox_inches="tight")
plt.show()

# Visualises the accident severity by road type
road_severity_melted = road_severity.melt(
    id_vars="Road_Type",
    value_vars=["Fatal_Count", "Serious_Count", "Slight_Count"],
    var_name="Severity",
    value_name="Count"
)

plt.figure(figsize=(12, 8))
for severity in ["Fatal_Count", "Serious_Count", "Slight_Count"]:
    subset = road_severity_melted[road_severity_melted["Severity"] == severity]
    plt.bar(
        subset["Road_Type"],
        subset["Count"], #allows it to visually be presented as a count
        label=severity.replace("_Count", ""),
        alpha=0.7
    )
#plots the bar charts
plt.title("Accident Severity by Road Type", fontsize=14)
plt.xlabel("Road Type", fontsize=12)
plt.ylabel("Accident Count by Severity", fontsize=12)
plt.legend(title="Severity Level")
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.tight_layout()
plt.savefig("figure11.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#libraries used
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Query to group accidents by hour and count the amount of times it happens
#strftime gets the hours from the time and allows me to visualise it by just hour to make it simple and easier
#to read
query = """
SELECT strftime('%H', Time) as Hour, COUNT(*) as Accident_Count
FROM Accidents
WHERE Time NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
GROUP BY Hour
ORDER BY Hour;
"""
accidents_by_hour = pd.read_sql_query(query, conn)
# Converts the hour column to integer for proper sorting
accidents_by_hour["Hour"] = accidents_by_hour["Hour"].astype(int)

# Print the calculations
print(accidents_by_hour)

# Visualises the time series analysis and settings of how it is to be presented
plt.figure(figsize=(12, 6))
plt.plot(
    accidents_by_hour["Hour"], 
    accidents_by_hour["Accident_Count"], 
    marker="o", color="blue", label="Accident Count"
)
plt.title("Time Series: Accidents by The Hour of Day", fontsize=14)
plt.xlabel("Hour of the Day", fontsize=12)
plt.ylabel("Number of Accidents", fontsize=12)
plt.xticks(range(0, 24))
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("figure7.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#The peaks at which accidents occur are usually during the morning (Going to work, dropping school kids?)
#And early afternoon. mid afternoon and late afternoon are noteable times
```

```python
#libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Query to calculate accident counts for each month in 2015 and 2016
#strftime extracts just the year and month
#missing values excluded/filtered
query = """
SELECT strftime('%Y-%m', Date) as Month, COUNT(*) as Accident_Count
FROM Accidents
WHERE Date NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
GROUP BY Month
ORDER BY Month;
"""
accidents_by_month = pd.read_sql_query(query, conn)


# Convert the Month column to datetime datatype so it can go in right order
accidents_by_month["Month"] = pd.to_datetime(accidents_by_month["Month"])

# Prints calculations
print(accidents_by_month)

# Visualises the time series by Month
plt.figure(figsize=(12, 6))
plt.plot(
    accidents_by_month["Month"], 
    accidents_by_month["Accident_Count"], 
    marker="o", color="yellow", label="Accident Count"
)
#plots with preference to make it aesthetic 
plt.title("Accidents by Month", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Accidents", fontsize=12)
plt.grid(alpha=0.3)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig("figure6.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Accidents are higher during winter months as they have the highest peak
```

```python
#libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
#allows for me to do a 3d plot
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Query to get data
query = """
SELECT Time, Number_of_Casualties, Accident_Severity
FROM Accidents
WHERE Time NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
  AND Number_of_Casualties NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
  AND Accident_Severity NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated');
"""
data = pd.read_sql_query(query, conn)

# preprocessing the time column to convert it to hour of the day
data["Time"] = pd.to_datetime(data["Time"], format="%H:%M", errors="coerce").dt.hour
# ensures to remove rows with missing values after conversion
data = data.dropna()

# map accident severity to numeric values for color-coding
severity_mapping = {"Slight": 1, "Serious": 2, "Fatal": 3}
data["Severity_Code"] = data["Accident_Severity"].map(severity_mapping)

# 3D Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
# Scatter plot
sc = ax.scatter(
    data["Time"], 
    data["Number_of_Casualties"], 
    data["Severity_Code"], 
    c=data["Severity_Code"], 
    cmap="coolwarm", 
    s=50, 
    alpha=0.7
)
# Title and label settings
ax.set_title("3D Plot of Time, Casualties, and Accident Severity", fontsize=14)
ax.set_xlabel("Time (Hour of Day)", fontsize=12)
ax.set_ylabel("Number of Casualties", fontsize=12)
ax.set_zlabel("Accident Severity", fontsize=12)
ax.grid(True)

# Colour bar for Accident Severity
cbar = plt.colorbar(sc, ax=ax, shrink=0.5, aspect=10)
cbar.set_label("Accident Severity (1=Slight, 2=Serious, 3=Fatal)", fontsize=10)
#pshows plot
plt.tight_layout()
plt.show()
```

```python
#libraries used
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Query to calculate the average speed limit and Num of casualties grouped by Road Type
#filters out missing values
query = """
SELECT 
    Road_Type,
    AVG(Speed_limit) AS Avg_Speed_Limit,
    AVG(Number_of_Casualties) AS Avg_Number_of_Casualties
FROM Accidents
WHERE 
    Road_Type NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
    AND Speed_limit NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
    AND Number_of_Casualties NOT IN ('Data missing or out of range', 'Unknown', 'Unallocated')
GROUP BY Road_Type
ORDER BY Avg_Speed_Limit DESC;
"""
data = pd.read_sql_query(query, conn)


# Plotting
fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar plot for Av Speed Limit
sns.barplot(
    data=data,
    x="Road_Type",
    y="Avg_Speed_Limit",
    ax=ax1,
    color="blue",
    label="Avg Speed Limit",
    alpha=0.7
)

# This creates line plot for Av Num of casualties and sets preference
ax2 = ax1.twinx()
sns.lineplot(
    data=data,
    x="Road_Type",
    y="Avg_Number_of_Casualties",
    ax=ax2,
    color="red",
    label="Avg Number of Casualties",
    marker="o"
)

# Sets all the Titles and labels and preference size and colour
ax1.set_title("Average Speed Limit and Number of Casualties by Road Type", fontsize=16)
ax1.set_xlabel("Road Type", fontsize=12)
ax1.set_ylabel("Average Speed Limit", fontsize=12, color="blue")
ax2.set_ylabel("Average Number of Casualties", fontsize=12, color="red")
ax1.tick_params(axis="x", rotation=45)
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

#presents the plot
plt.tight_layout()
plt.savefig("figure13.png", dpi=300, bbox_inches="tight")
plt.show()
```
